import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import adjusted_rand_score

```python
# --------------------------------
# Load and prepare the data
# --------------------------------
df = pd.read_csv("../data/heart_disease_selected.csv")
X = df.drop("target", axis=1)
y_true = df["target"]

# Scaling for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## K-Means Clustering with Elbow Method

```python
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o')
plt.title("📉 Elbow Method for Optimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()

# From plot, choose k (say 2 since we have binary labels)
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

## Hierarchical Clustering + Dendrogram

```python
# Compute linkage matrix
link_matrix = linkage(X_scaled, method='ward')

# Dendrogram
plt.figure(figsize=(12, 5))
dendrogram(link_matrix, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=12., show_contracted=True)
plt.title("🌳 Dendrogram (Hierarchical Clustering)")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()

# Apply Hierarchical Clustering (cut at 2 clusters)
hc = AgglomerativeClustering(n_clusters=2)
hc_labels = hc.fit_predict(X_scaled)
```

## Compare clusters with actual labels

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# تحويل إلى DataFrame للرسم
pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["KMeans"] = kmeans_labels
pca_df["Hierarchical"] = hc_labels
pca_df["Actual"] = y_true.values
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="KMeans", palette="Set2", s=60)
plt.title("🎯 KMeans Clustering (PCA Visualization)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="Hierarchical", palette="Set1", s=60)
plt.title("🌿 Hierarchical Clustering (PCA Visualization)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(data=pca_df, x="PC1", y="PC2", hue="Actual", palette="coolwarm", s=60)
plt.title("✅ Actual Labels (PCA Visualization)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

# Adjusted Rand Index: measures similarity between two clusterings (range -1 to 1)
kmeans_ari = adjusted_rand_score(y_true, kmeans_labels)
hc_ari = adjusted_rand_score(y_true, hc_labels)

print(f"🔁 Adjusted Rand Index (K-Means): {kmeans_ari:.3f}")
print(f"🔁 Adjusted Rand Index (Hierarchical): {hc_ari:.3f}")

# Confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(confusion_matrix(y_true, kmeans_labels), annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("🔷 K-Means vs True Labels")
axes[0].set_xlabel("Cluster Label")
axes[0].set_ylabel("Actual Label")

sns.heatmap(confusion_matrix(y_true, hc_labels), annot=True, fmt="d", cmap="Greens", ax=axes[1])
axes[1].set_title("🌿 Hierarchical vs True Labels")
axes[1].set_xlabel("Cluster Label")
axes[1].set_ylabel("Actual Label")

plt.tight_layout()
plt.show()
```
