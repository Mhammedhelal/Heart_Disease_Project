```python
import pandas as pd
```

#### Load Dataset

```python
data = pd.read_csv('../data/heart_disease.csv')
data.head()
```

#### Handle Missing Values

```python
data
```

```python
data.dropna()
```

##### No Missing Data

#### Encode Categorical Features

```python
categorical_cols = ["cp", "restecg", "slope", "thal"]
data_encoded = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
data_encoded.replace({"True": 1, "False": 0}, inplace=True)
data_encoded.to_csv("../data/heart_disease_encoded.csv", index=False)
data_encoded.head()
```

#### Scale Numeric Features

```python
from sklearn.preprocessing import StandardScaler
```

```python
numeric_cols = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]
scaler = StandardScaler()
data_encoded[numeric_cols] = scaler.fit_transform(data_encoded[numeric_cols])
data_encoded.to_csv("../data/heart_disease_scaled.csv", index=False)
```

```python
data_encoded.head()
```

#### EDA: Visualize using histograms, boxplots, correlation heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.set(style="whitegrid", palette="muted", font_scale=1.1)
plt.rcParams["figure.figsize"] = (10, 6)


# ========== 1. Histograms ==========
data_encoded.hist(bins=30, figsize=(18, 12), color="#007acc", edgecolor="black")
plt.suptitle("🔍 Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()

# ========== 2. Boxplots ==========
numeric_cols = ["age", "trestbps", "chol", "thalach", "oldpeak", "ca"]
for col in numeric_cols:
    plt.figure()
    sns.boxplot(x='target', y=col, data=data_encoded)
    plt.title(f"📦 Boxplot of {col} vs Target")
    plt.show()

# ========== 3. Correlation Heatmap ==========
plt.figure(figsize=(14, 10))
corr = data_encoded.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)
plt.title("🔥 Correlation Heatmap")
plt.show()
```
